```python
#過濾不是垃圾信的訊息
ham = data[data['status'] == 'ham']
ham
```

```python
#過濾不是垃圾訊息中，有ＯＫ的訊息
mask = ham['msg'].str.contains('OK')
mask
```

```python
ham[mask]
```
