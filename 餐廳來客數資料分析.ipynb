```python
import pandas as pd

res = pd.read_csv('https://raw.githubusercontent.com/Code-Gym/python-dataset/master/RestaurantVisitors.csv')
res
```

```python
res.index = pd.to_datetime(res['date'])
res.head(7)
```

```python
res['week'] = res.index.week
```

```python
res.groupby('week').rest1.sum()#餐廳1的每周來客數
```

```python
res[(res.holiday == 1)|(res.weekday == 'Saturday')|(res.weekday == 'Sunday')].loc[:,'rest1':'rest4'].mean()#假日四間餐廳平均來客數
```

```python
res[res.holiday != 1].loc[:,'rest1':'rest4'].mean()#非假日四間餐平均來客數
```
