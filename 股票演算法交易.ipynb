```python
import sys,requests,json

date=sys.argv[1]
stock_symbol=sys.argv[2]

html = requests.get('https://www.twse.com.tw/exchangeReport/BWIBBU?response=json&date='+date+'&stockNo='+stock_symbol)

if html.status_code == requests.codes.ok:
    jcontent=json.loads(html.text)
    data=jcontent['data']
    print(data)
else:
    print('失敗')
```
