```python
import pandas as pd 

df1 = pd.DataFrame({'A':['A0', 'A1', 'A2', 'A3'],
                    'B':['B0', 'B1', 'B2', 'B3'],
                    'C':['C0', 'C1', 'C2', 'C3'],
                    'D':['D0', 'D1', 'D2', 'D3'],},
                  index = [0, 1, 2, 3])
                    
df2 = pd.DataFrame({'A':['A4', 'A5', 'A6', 'A7'],
                    'B':['B4', 'B5', 'B6', 'B7'],
                    'C':['C4', 'C5', 'C6', 'C7'],
                    'D':['D4', 'D5', 'D6', 'D7'],},
                  index = [4, 5, 6, 7])
df1.append(df2)
```

```python
df3 = pd.DataFrame({'A':['A8', 'A9', 'A10', 'A11'],
                    'B':['B8', 'B9', 'B10', 'B11'],
                    'C':['C8', 'C9', 'C10', 'C11'],
                    'D':['D8', 'D9', 'D10', 'D11'],},
                  index = [8, 9, 10, 11])
df1.append([df2,df3])
```

```python
df4 = pd.DataFrame({'B':['B2', 'B3', 'B6', 'B7'],
                    'F':['C2', 'C3', 'C6', 'C7'],
                    'D':['D2', 'D3', 'D6', 'D7'],},
                  index = [2, 3, 6, 7])
df1.append(df4)
```

```python
df1.append(df4, ignore_index=True)
```

```python
df1.loc[4] = ['A4', 'B4', 'C4', 'D4']
df1
```

```python
df1 = df1.append({'A':'A5', 'B':'B5', 'C':'C5', 'D':'D5'}, ignore_index=True)
df1
```

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df4], axis=1, join='outer', sort=False)#橫向聯集
```

```python
pd.concat([df1, df4], axis=1, join='inner', sort=False)#橫向交集
```

```python
pd.concat([df1, df4], axis=0, join='inner', sort=False)#縱向交集
```

```python
result = pd.concat([df1, df2, df3], keys=['x', 'y', 'z'])#keys=['次索引值１', '次索引值2', .... ]
result
```

```python
result.loc['y']#尋找資料
```

```python
left = pd.DataFrame({'Key':['K0', 'K1', 'K2', 'K3'],
                    'A':['A0', 'A1', 'A2', 'A3'],
                    'B':['B0', 'B1', 'B2', 'B3'],})


right = pd.DataFrame({'Key':['K0', 'K1', 'K2', 'K3'],
                    'C':['C0', 'C1', 'C2', 'C3'],
                    'D':['D0', 'D1', 'D2', 'D3'],})

left.merge(right, on='Key')#輸入Ｋey是指定特定欄位結合,也可以不輸入
```

```python
apple = pd.DataFrame({'Key1':['K0', 'K0', 'K1', 'K2'],
                    'Key2':['K0', 'K1', 'K0', 'K1'],
                    'A':['A0', 'A1', 'A2', 'A3'],
                    'B':['B0', 'B1', 'B2', 'B3'],})


orange = pd.DataFrame({'Key1':['K0', 'K1', 'K1', 'K2'],
                    'Key2':['K0', 'K0', 'K0', 'K0'],
                    'C':['C0', 'C1', 'C2', 'C3'],
                    'D':['D0', 'D1', 'D2', 'D3'],})

```

```python
apple
```

```python
orange
```

```python
apple.merge(orange, on=['Key1', 'Key2'])#指定key1跟key2欄位有相同的合併，但有一筆資料key相同後面的值不同,所以要補上一行
```

```python
apple.merge(orange, how='left', on=['Key1', 'Key2'])#以apple為主的交集且包含apple
```

```python
apple.merge(orange, how='right', on=['Key1', 'Key2'])#以orange為主的交集且包含orange
```

```python
apple.merge(orange, how='outer', on=['Key1', 'Key2'])#聯集
```

```python
apple.merge(orange, how='inner', on=['Key1', 'Key2'])#交集
```

```python
apple.merge(orange, how='outer', on=['Key1', 'Key2'], indicator=True)#找出資料位置在哪裡,both都有,left在左邊,right在右邊
```

```python
apple.merge(orange, on=['Key1'])#因為沒有設定key2,所以系統自動把他分成兩個
```

```python
apple.merge(orange, on=['Key1'], suffixes=('_apple', '_orange'))#自己定義後面的欄位名稱
```
