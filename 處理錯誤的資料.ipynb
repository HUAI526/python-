```python
import pandas as pd

iris = pd.read_csv('https://raw.githubusercontent.com/Code-Gym/python-dataset/master/iris.csv',
                  names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
iris.head()
```

```python
del iris['class']#刪除欄位
iris.head()
```

```python
#檢查欄位中是否有空值
iris.isnull().sum()
```

```python
#建立空值
import numpy as np
iris.iloc[1:4, 1:2] = np.nan
iris.head()
```

```python
iris.isnull().sum()
```

```python
#將空值替換成０
iris.fillna(0, inplace=True)

iris.head()
```

```python
iris.iloc[1:4, 1:2] = np.nan
iris.head()
```

```python
#刪除資料中的空值，直的刪除
iris = iris.dropna(axis = 'columns', how='any')
iris.head()
```

```python
##刪除資料中的空值，橫的刪除
iris = iris.dropna(axis = 'index', how='any')
iris.head()
```

```python
#重置排序號碼
iris = iris.reset_index(drop=True)
iris.head()
```
